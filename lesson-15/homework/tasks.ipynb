```python
#task2
import numpy as np
import matplotlib.pyplot as plt

x=np.linspace(0,2,10000)
def f_x(x):
    return x**2 - 4*x +4
y=f_x(x)
plt.plot(x,y, label = "$f(x) = x^2 - 4x + 4$", color="black")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()


thetha=np.linspace(0,2,10000)
def sin_x(thetha):
    return np.sin(thetha)
def cos_x(thetha):
    return np.cos(thetha)

y2=sin_x(thetha)
y3=cos_x(thetha)
plt.plot(thetha, y2, label="$g(x) = sin(x)$", color='blue', linestyle="--")
plt.plot(thetha, y3, label="$h(x) = cos(x)$", color = 'red', linestyle="-.")
plt.legend()
plt.show()
```

```python
#task3

import matplotlib.pyplot as plt
import numpy as np

fig, axes = plt.subplots(2,2, figsize=(8,8))
x=np.linspace(-2,2,10000)

axes[0,0].plot(x, x**3, label="$y1 = x^3$", color = 'blue')

axes[0,1].plot(x, np.sin(x), label="y2=sinx", color="red", linestyle="--")

axes[1,0].plot(x, np.exp(x), label="$y3=e^x$", color="green", linestyle="-.")
x_pos = x[x > -1]
axes[1,1].plot(x_pos, np.log(x_pos), color = 'black',label ="$ y4 = log (x+1) $", linestyle= ":")

plt.legend()
plt.xlabel("X")
plt.ylabel("Y")
plt.tight_layout()
plt.show()

```

```python
#task6

import numpy as np
import matplotlib.pyplot as plt

fig=plt.figure()
ax=plt.axes(projection='3d')

x=np.linspace(-5,5,1000)
y=np.linspace(-5,5,1000)
xx, yy = np.meshgrid(x,y)
zz=np.cos(xx**2 + yy**2)
ax.plot_surface(xx,yy,zz, cmap="viridis", label="$f(x,y) = cos(x^2 + y^2)")


ax.legend()
```

```python
#task8
import matplotlib.pyplot as plt
import numpy as np

# Product names (x-axis labels)
products = ["A", "B", "C", "D"]

# Data: quantities in different seasons
season_1 = np.array([10, 6, 8, 5])
season_2 = np.array([8, 9, 6, 7])
season_3 = np.array([12, 4, 7, 8])
season_4 = np.array([7, 5, 6, 6])

# Plot each season, stacking them
fig, ax = plt.subplots(figsize=(8, 6))

bar1 = ax.bar(products, season_1, label="Season 1", color='blue')
bar2 = ax.bar(products, season_2, bottom=season_1, label="Season 2", color='red')
bar3 = ax.bar(products, season_3, bottom=season_1 + season_2, label="Season 3", color='green')
bar4 = ax.bar(products, season_4, bottom=season_1 + season_2 + season_3, label="Season 4", color='purple')

# Adding labels for each section
for i, product in enumerate(products):
    total_height = 0  # Track the bottom position
    for season, color in zip([season_1, season_2, season_3, season_4], ['blue', 'red', 'green', 'purple']):
        height = season[i]
        if height > 0:
            ax.text(i, total_height + height / 2, str(height), ha='center', va='center', fontsize=10, color='white')
        total_height += height  # Update bottom position

# Labels and title
ax.set_ylabel("Total Quantity")
ax.set_xlabel("Products")
ax.set_title("Stacked Bar Chart of Product Quantities Over 4 Seasons")
ax.legend()
plt.show()
```
