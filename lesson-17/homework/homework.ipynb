```python
#2

path ="../data/employee.csv"

df_csv=pd.read_csv(path)
def normalize(x):
    return (x - x.mean()) / (x.max() - x.min()) #normalization

df_csv["NORMALIZED_SALARY"] = df_csv.groupby("DEPARTMENT")["BASE_SALARY"].transform(normalize) #applying function

df_csv.head()
```

```python
#3

path ="../data/movie.csv"
df_csv=pd.read_csv(path)

def classify(x):
    if pd.isna(x):  # Handle missing values
        return "Unknown" #hangling Nan value
    elif x<60:
        return "Short"
    elif x>=60 and x<120:
        return "Medium"
    else:
        return "Long"
    
df_csv["Class"]=df_csv["duration"].apply(classify)
df_csv.head()
```

```python
#task 3

import pandas as pd

# Load the dataset
path = "../data/titanic.xlsx"
df = pd.read_excel(path)

# Define functions for the pipeline
def filter_survivors(data):
    return data[data["Survived"] == 1]

def fill_missing_age(data):
    data["Age"] = data["Age"].fillna(data["Age"].mean())
    return data

def create_fare_per_age(data):
    data["Fare_Per_Age"] = data["Fare"] / data["Age"]
    return data

# Apply the pipeline using pipe()
df_cleaned = (df.pipe(filter_survivors)
                .pipe(fill_missing_age)
                .pipe(create_fare_per_age))

# Display result
df_cleaned.head()
```

```python
#2

import pandas as pd

# Load the dataset
path = "../data/flights"
df = pd.read_parquet(path)

# Define pipeline functions
def filter_delayed_flights(data):
    return data[data["DepartureDelay"] > 30]

def calculate_delay_per_hour(data):
    data["Delay_Per_Hour"] = data["DepartureDelay"] / data["ScheduledDuration"]
    return data

# Apply the pipeline using pipe()
df_cleaned = (df.pipe(filter_delayed_flights)
                .pipe(calculate_delay_per_hour))

# Display result
print(df_cleaned.head())
```
