```python
#part 3

#1
import pandas as pd
path = '../data/iris.json' #file on folder

df_json = pd.read_json(path) #reading the file

# Select only numerical columns
numerical_df = df_json.select_dtypes(include=["number"]) #since first columns is string, I should only choose numeric values
mean=numerical_df.mean()    #calculating mean
median=numerical_df.median() #calculating median
std = numerical_df.std() #calculating std

print("Mean: ", mean, "-"*50)
print("Median: ", median, "-"*50)
print("Standard deviation: ", std, "-"*50)
```

```python
#2
import pandas as pd

path = "../data/titanic.xlsx"
df_excel=pd.read_excel(path) #reading excel file

df_age=df_excel['Age'] #finding row named age
min=df_age.min()
max=df_age.max()
sum=df_age.sum()

print(f"Minium value: {min}", "-"*50)
print(f"Maximum value: {max}", "-"*50)
print(f"Sum: {sum}", "-"*50)
```

```python
#3

import pandas as pd
path ="../data/movie.csv"

df_csv=pd.read_csv(path)

# Identify the director with the highest total director_facebook_likes
top_director = df_csv.groupby("director_name")["director_facebook_likes"].sum().idxmax()

# Find the 5 longest movies and their respective directors
longest_movies = df_csv[["movie_title", "duration", "director_name"]].nlargest(5, "duration")

# Display results
print("Director with the highest total Facebook likes:", top_director)
print("\n5 Longest Movies and Their Directors:")
print(longest_movies)
```

```python
#4

import pandas as pd
path ="../data/fflights" 
df_parquet=pd.read_parquet(path)

missing_values=df_parquet.is_null().sum()
print("Missing Value: ", missing_values, "\n")
# Fill missing values in numerical columns with the column mean
df_parquet.fillna(df_parquet.mean(numeric_only=True), inplace=True)

# Verify missing values are filled
print("\nMissing values after filling:\n", df_parquet.isnull().sum())
```
